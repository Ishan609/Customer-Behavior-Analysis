```python
df["Review Rating"] = df["Review Rating"].fillna(
    df.groupby(["Item Purchased", "Location"])["Review Rating"]
      .transform("median")
)
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df.columns
```

```python
# creating age group
labels=['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group']= pd.qcut(df['age'],q=4, labels=labels)
```

```python
df[['age', 'age_group']]
```

```python
# Creating column purchase frequency days 
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
